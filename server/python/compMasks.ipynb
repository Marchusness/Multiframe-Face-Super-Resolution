```python
from sewar.full_ref import mse, rmse, psnr, uqi, ssim, ergas, scc, rase, sam, msssim, vifp
import cv2 as cv
import os

m1 = []
r2 = []
p3 = []
u4 = []
s5 = []
e6 = []
s7 = []
r8 = []
s9 = []
m10 = []
v11 = []

base = cv.imread("./1000.png")
for folder in ["32", "64", "128", "256", "320", "640", "1280", "2560"]:
    cm1 = []
    cr2 = []
    cp3 = []
    cu4 = []
    cs5 = []
    ce6 = []
    cs7 = []
    cr8 = []
    cs9 = []
    cm10 = []
    cv11 = []

    folderPath = "./outputMasks/" + folder
    for img in ["2", "5", "20", "100", "1000"]:
        cur = cv.imread(folderPath + "/" + img + ".png")
        cm1.append(mse(base, cur))
        cr2.append(rmse(base, cur))
        cp3.append(psnr(base, cur))
        cu4.append(uqi(base, cur))
        cs5.append(ssim(base, cur))
        ce6.append(ergas(base, cur))
        cs7.append(scc(base, cur))
        cr8.append(rase(base, cur))
        cs9.append(sam(base, cur))
        cm10.append(msssim(base, cur))
        cv11.append(vifp(base, cur))

    m1.append(cm1)
    r2.append(cr2)
    p3.append(cp3)
    u4.append(cu4)
    s5.append(cs5)
    e6.append(ce6)
    s7.append(cs7)
    r8.append(cr8)
    s9.append(cs9)
    m10.append(cm10)
    v11.append(cv11)


```
